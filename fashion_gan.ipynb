## 1. Dataset Preparation

```python
# First, lets set up our imports, dataset, and DataLoader.

#Imports
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as T
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
import statistics

#Dataset downloading and prep
torch.manual_seed(42)
transform = T.Compose([
    T.ToTensor(),
    T.Normalize(0.5,0.5)
])

train_set = torchvision.datasets.FashionMNIST(
    root='.',
    train=True,
    download=True,
    transform=transform
)

#Dataloader setup
batch_size = 32
train_loader = torch.utils.data.DataLoader(
    train_set,
    batch_size=batch_size,
    shuffle=True
)

#Plot all the images of the first batch in grid
images, labels = next(iter(train_loader))
grid = make_grid(0.5-images/2, 8, 4)
plt.imshow(grid.numpy().transpose((1,2,0)), cmap="gray_r")
plt.axis("off")
plt.show()
```

## 2. GAN Architecture

```python
#Generator Architecture 
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.z_dim = 100
        self.model = nn.Sequential(
            nn.Linear(self.z_dim, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 1024),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(1024, 784),
            nn.Tanh()
        )
    
    def forward(self, z):
        img = self.model(z)
        img = img.view(img.size(0), 784)
        return img

#Discriminator Architecture 
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        
        self.model = nn.Sequential(
                nn.Linear(784, 1024),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Dropout(0.3),
                nn.Linear(1024, 512),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Dropout(0.3),
                nn.Linear(512, 256),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Dropout(0.3),
                nn.Linear(256, 128),
                nn.LeakyReLU(0.2, inplace=True),
                nn.Dropout(0.3),
                nn.Linear(128, 1),
                nn.Sigmoid()
        )
        
    def forward(self, img):
        img_flat = img.view(img.size(0), -1)
        validity = self.model(img_flat)
        return validity
    
#Setting device
if torch.cuda.is_available():
    device = torch.device('cuda:0')
else:
    device = torch.device('cpu')

#Creating models and sending to device
generator  = Generator().to(device)
discriminator = Discriminator().to(device)
```

## 3. Loss and Optimizer

```python
#Learning rate, we use .0001. This gave me the best results in my testing.
lr = 2e-4

#Loss function, we use Binary Cross Entropy Loss
loss_fn = nn.BCELoss()

#Optimizers, we use Adam
optim_g = torch.optim.Adam(generator.parameters(), lr = lr)
optim_d = torch.optim.Adam(discriminator.parameters(), lr = lr)
```

## 4. Training Loop

```python
#Loss avgs for each epoch for the graph.
g_losses = []
d_losses = []

epochs = 50
for epoch in range(epochs):
    # For each epoch, iterate over the DataLoader of the training dataset
    for i, (real_images, _) in enumerate(train_loader):
        # For each batch, perform the following steps:
        current_batch_size = real_images.size(0)
        real_images = real_images.to(device)
        # Identity matrix for real and fake labels.
        real_labels = torch.ones(current_batch_size, 1).to(device)
        fake_labels = torch.zeros(current_batch_size, 1).to(device)
        optim_d.zero_grad()
        # Learn on discriminator on real images.
        real_outputs = discriminator(real_images)
        d_loss_real = loss_fn(real_outputs, real_labels)
        # Generate fake images.
        z = torch.randn(current_batch_size, 100).to(device)
        fake_images = generator(z)
        # Learn on discriminator on fake images.
        fake_outputs = discriminator(fake_images.detach())
        d_loss_fake = loss_fn(fake_outputs, fake_labels)
        d_loss = (d_loss_real + d_loss_fake) / 2
        # Discriminator backprop.
        d_loss.backward()
        optim_d.step()
        # Generator learning.
        optim_g.zero_grad()
        fake_outputs = discriminator(fake_images)
        g_loss = loss_fn(fake_outputs, real_labels)
        # Generator backprop.
        g_loss.backward()
        optim_g.step()
        
        # Add losses for later plotting.
        g_losses.append(g_loss.item())
        d_losses.append(d_loss.item())
        if (i+1) % 500 == 0:
            #Print progess to make sure training is not stuck.
            print(f'Epoch ({epoch+1}/{epochs})  Iter ({i+1}/{len(train_loader)})')
        if i+1 == len(train_loader) or (epoch == 0 and i == 0):
            if (epoch == 0 and i == 0):
                print("First iteration of first epoch (Noise):")
            if i+1 == len(train_loader):
                print(f'Epoch ({epoch+1}/{epochs})  Iter ({i+1}/{len(train_loader)})  d_loss: {d_loss.item():.4f}  g_loss: {g_loss.item():.4f}')
            #Plot generated images at the end of each epoch as a check in, and the first iteration to show noise.
            z = torch.randn(9, 100).to(device)
            sample_images = generator(z)
            sample_images = sample_images.view(sample_images.size(0), 28, 28).unsqueeze(1).data.cpu()
            grid = make_grid(sample_images, nrow=3, normalize=True).permute(1,2,0).numpy()
            plt.imshow(grid)
            plt.axis('off')
            plt.show()
        
    
    #Saving models every 10 epochs. Saving discriminator isn't necessarily helpful, as all you really need is the generator. 
    if (epoch + 1) % 10 == 0:
        gPATH = "./fashion_gen_"+str(epoch)+".pth"
        torch.save(generator.state_dict(), gPATH)
        dPATH = "./fashion_dis_"+str(epoch)+".pth"
        torch.save(generator.state_dict(), dPATH)
```

## 5. Visualizations

```python
#Plotting loss curves.
plt.figure(figsize=(10, 5))
plt.plot(g_losses, label='Generator Loss')
plt.plot(d_losses, label='Discriminator Loss')
plt.ylim(-1,10)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Generator and Discriminator Loss Curves')
plt.legend()
plt.grid(True)
plt.show()
```

## - Model Architecture Choices
For the architecture of my GAN, I mainly followed what was outlined on the slides, without going into the extra details of possible suggestions. I employed a standard fully connected feed forward netowrk, with no convolution layers for both the generator and the discriminator. My generator starts off with the z_dim of 100, coming from the size I set for the noise input, increasing each layer to 1024. Once the layer size hits 1024, I brought it back down to be 784 which is the size of a flattened 28x28 image. In between each layer as suggested I used LeakyReLU as my activation function, since it allows for there to be negative values compared to the standard ReLU. My discriminator architecture is the inverse of my generator architecture, however the last layer ends with a size of 1 so that we can obtain a classification of real/fake as the output. Lastly we end with a sigmoid, as this limits our values to between 0 and 1, for real and fake. I additionally included some dropout layers, which I am not completely sure if it was necessary to get it working, but it theoretically would help to avoid overfitting for the model. I used a learning rate of .0002 (2e-4) and a batch size of 32, as was suggested in class, and I did find that this worked the best for me. 
## - Training Process
My training loop has 50 epochs, training on the FashionMNIST dataset. In my outputs, I outputed the first images generated by the generator, showing pure noise. However, the generator learns very quickly with rough outlines of images starting to appear around the fourth epoch. The losses appear to level off fairly quickly, with them hovering around 0.65 for the discriminator and 0.9 for the generator. The overall structure for the training loop is as follows: For each epoch, iterate over the DataLoader of the training dataset, For each epoch, iterate over the DataLoader of the training dataset, Learn on discriminator on real images, Generate fake images, Learn on discriminator on fake images, Discriminator backprop, Generator learning, Generator backprop.
